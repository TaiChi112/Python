## Part 1: Understanding Digital Image Processing

### What is Digital Image Processing?
Digital image processing involves using algorithms and mathematical models to process and analyze digital images. The goal is to enhance image quality, extract meaningful information, and automate image-based tasks

### The Objective of Image Processing
The primary objectives of image processing are:
1. **Visualization**: Observe objects that are not visible.
2. **Image sharpening and restoration**: Improve the quality of an image.
3. **Image retrieval**: Seek for the image of interest.
4. **Measurement of pattern**: Measures various objects in an image.
5. **Image recognition**: Distinguish the objects in an image.

### The basic steps in digital image processing are:
1. **Image acquisition:** Capturing an image digitally or importing an existing image.



2. **Image enhancement: ** Improving visual quality (contrast, noise reduction, etc.)



3. **Image restoration:** Removing degradation (blur, noise, distortion).



4. **Image segmentation:** Dividing the image into regions corresponding to objects.



5. **Image representation & description:** Representing the image for computer analysis.



### Digital Image Types
The main types of digital images are:
- <b>Color images</b> Images representing color, usually as three separate Red/Green/Blue (RGB) channels. A 24-bit RGB color image has 8 bits per channel. Other color spaces like HSV/HSL are also used.



- <b>Grayscale images</b> Monochrome images containing brightness information only, no color. Typically 8 bits/pixel allowing 256 different gray levels. Used widely as they contain enough information for many tasks while being - computationally simpler than color.



- <b>Binary images:</b> The simplest images with only two pixel values - black (0) and white (1). Used when only general shape/outline information is needed.



<b>Multispectral images:</b> Images with data outside the normal visible range, such as infrared, ultraviolet, X-ray etc. The spectral bands are mapped to visual colors for display.

### Digital Image Coordinate System
A digital image can be represented as a 2D function f(x,y) where x and y are spatial coordinates and the value of f is the intensity at that point. The coordinates are arranged in a matrix with the origin (0,0) at the top-left



Here the image has dimensions MxN pixels.

### Basic Image Processing Operations
Now let's see how to perform some basic operations on images using Python and OpenCV.
1. <b>Reading</b> an Image
2. <b>Resizing</b> an Image

This covers some of the basic image processing operations using OpenCV in Python. Digital image processing has a wide range of applications in computer vision, medical imaging, remote sensing, multimedia and more.

```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Opening and Reading an Image
image_path = 'lena.png'
image = cv2.imread(image_path)

print(image.shape, type(image))
# Resizing
resized_image = cv2.resize(image, (500, 400))  # Resize to 500x400 pixels

# Converting to Grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Displaying Images
plt.figure(figsize=(10, 5))

plt.subplot(121)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for display
plt.title('Original Image')

plt.subplot(122)
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')

plt.show()
```

Here's a comparison of basic image processing operations using OpenCV, scikit-image, and PIL (Pillow), along with code examples and image format conversions:
#### 1.Read Image


```python
import cv2
from skimage import io
from PIL import Image

# OpenCV
img_cv = cv2.imread('image.jpg')  # Reads in BGR format

# scikit-image
img_ski = io.imread('image.jpg')

# PIL
img_pil = Image.open('image.jpg')  
```

#### 2. Write Image

```python
# OpenCV
cv2.imwrite('output_cv.jpg', img_cv)

# scikit-image
io.imsave('output_ski.jpg', img_ski)

# PIL
img_pil.save('output_pil.jpg')
```

```python
import numpy as np

# OpenCV
resized_cv = cv2.resize(img_cv, (500, 300))

# scikit-image
resized_ski = io.resize(img_ski, (300, 500))

# PIL
resized_pil = img_pil.resize((500, 300))
```
